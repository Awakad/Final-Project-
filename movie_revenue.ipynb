```python
# Import our dependencies
import pandas as pd
import numpy as np
```

```python
# Load the movie_metadata.csv dataset.
file_path = "movie_metadata.csv"
moviedata = pd.read_csv(file_path, index_col=0)
moviedata
```

```python
# Check columns
moviedata.columns
```

```python
# Check data type
moviedata.dtypes
```

```python
# select the revelant columns
moviedata_df = moviedata[["director_name", "num_critic_for_reviews", "duration", "budget", "gross", "genres", "imdb_score" ]]
moviedata_df
```

```python
# Remove the "genres" column. 
moviedata_df = moviedata_df.drop(['genres'], axis=1)
moviedata_df.head()
```

```python
# Remove rows that have at least 1 null value.
moviedata_df = moviedata_df.dropna()
moviedata_df
```

```python
# Keep the rows where coins are mined.
moviedata_df = moviedata_df[moviedata_df['imdb_score'] > 5.0]
moviedata_df
```
